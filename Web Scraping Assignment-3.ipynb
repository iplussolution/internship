```python
#6. Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) from trak.in.

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for the funding deals page
url = "https://trak.in/india-startup-funding-investment-2015/"

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content)

# Find the table containing the funding deals
table = soup.find("table", class_="tablepress tablepress-id-48")

# Create empty lists to store the scraped data
dates = []
startups = []
industries = []
sub_verticals = []
cities = []
investors = []
investment_types = []
amounts = []

# Find all the rows in the table (excluding the header row)
rows = table.find("tr")[1:]

# Iterate over each row and extract the data
for row in rows:
    # Extract the columns for each row
    columns = row.find_all("td")

    # Extract the data from the columns
    date = columns[0].text.strip()
    startup = columns[1].text.strip()
    industry = columns[2].text.strip()
    sub_vertical = columns[3].text.strip()
    city = columns[4].text.strip()
    investor = columns[5].text.strip()
    investment_type = columns[6].text.strip()
    amount = columns[7].text.strip()

    # Check if the date is within the specified range (Jan 21 - March 21)
    if "Jan" in date or "Feb" in date or "Mar" in date:
        dates.append(date)
        startups.append(startup)
        industries.append(industry)
        sub_verticals.append(sub_vertical)
        cities.append(city)
        investors.append(investor)
        investment_types.append(investment_type)
        amounts.append(amount)

# Create a dataframe with the scraped data
data = {
    "Date": dates,
    "Startup": startups,
    "Industry": industries,
    "Sub Vertical": sub_verticals,
    "City": cities,
    "Investor": investors,
    "Investment Type": investment_types,
    "Amount": amounts,
}
df = pd.DataFrame(data)

# Save the dataframe to a CSV file
df.to_csv("funding_deals.csv", index=False)

print("Scraping and saving complete.")
```
